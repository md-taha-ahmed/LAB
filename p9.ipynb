#### 9- implement the non-parametric Locally Weighted ***Regression algorithm*** in order to fit data points. Select appropriate data set for your experiment and draw graphs

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import csv
def regline(x,y):
    plt.bar(x,y,label = "AMCEC admission")
    plt.xlabel('year')
    plt.ylabel('nos')
    plt.scatter(x,y,c='red')
    plt.show()
    mx=np.mean(x)
    my=np.mean(y)
    n=len(x)
    up=0
    M=0
    dw=0
    for i in range(n):
        up+=(x[i]-mx)*(y[i]-my)
        dw+=(x[i]-mx)**2
    M=up/dw
    c=my-(M*mx)
    print("linear regression slope =",M)
    print("linear regression constant =",c)
    max_x=np.max(x)+1
    min_x=np.min(x)-1
    #print (max_x)
    x1=0
    x1=np.linspace(min_x,max_x,6)
    y1=M*x1+c
    plt.plot(x1,y1,color='blue')
    print(x)
    plt.scatter(x,y,c='red')
    plt.show()
    print( "Enter which year admission prediction")
    year = int(input())
    adm = M*year+c
    print("Predicted admission =",adm)
def main():
    file=r"data/AMCEC.csv"
    data=pd.read_csv(file)
    display(data)
    x=data['year'].values
    y=data['nos'].values
    regline(x,y)
    
main()
```
